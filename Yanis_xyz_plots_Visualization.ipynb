```python
# Loading the data and exporting them into a list

from math import *
import matplotlib.pyplot as plt
%matplotlib inline
import statistics
from scipy.signal import savgol_filter
import numpy as np
```

```python
def find_loss(mot,k,l,path):
    x_hat=0
    x_hat_str=0
    x_hat_list=[]
    x_gt=0
    x_gt_str=0
    x_gt_list=[]
    loss_avg=[]    
    liste = []
    
    fh = open(path)
    read_lines = fh.readlines()

    for line in read_lines:
        liste.append(line)
    fh.close()
    
    if mot=='x-hat':
        mot1= 'x-hat'
        mot2= 'q-hat'
    if mot=='x-gt':
        mot1= 'x-gt'
        mot2= 'q-gt'
    idx=0
    for i in range(0,len(liste),1):    
        
        if mot1 in liste[i]:
            idx+=1
            #print(liste[i])
            x_hat=liste[i].split("\t")[-1]
            #print(' x_hat', x_hat)
            x_hat_str=x_hat.split()[-k]
            #print('x_hat_str',x_hat_str)
            x_hat_str=x_hat_str.replace("]","")
            x_hat_str=x_hat_str.replace("[","")
            x_hat_str=x_hat_str.split(",")[-l]
            #print('x_hat_str',x_hat_str)
            x_hat_list.append(float(x_hat_str))
            
        if mot2 in liste[i]:
            idx+=1
            x_gt=liste[i].split("\t")[-1]
            x_gt_str=x_gt.split()[-k] 
            x_gt_str=x_gt_str.replace("]","")
            x_gt_str=x_gt_str.replace("[","")
            x_gt_str=x_gt_str.split(",")[-l]
            x_gt_list.append(float(x_gt_str))

    #print(idx)
    return x_hat_list , x_gt_list

```

```python
# train --> X_pred vs X_gt
signal_hat=find_loss('x-hat',3,4,'/Users/bouadiyanis/DeepLIO_Test_ter/outputs/train_deeplio_20210620_163905/train_deeplio_20210620_163905.log')
signal_gt=find_loss('x-gt',3,4,'/Users/bouadiyanis/DeepLIO_Test_ter/outputs/train_deeplio_20210620_163905/train_deeplio_20210620_163905.log')


signal_hat = savgol_filter(signal_hat[0], 21, 3)
signal_gt = savgol_filter(signal_gt[0], 21, 3)


plt.figure(figsize=(12, 4), dpi=80)
plt.plot(signal_hat, 'b', label='x_prediction')
plt.plot(signal_gt, 'r',label='x_gt')
plt.legend()
print(len(signal_hat))
print(len(signal_gt))

```
