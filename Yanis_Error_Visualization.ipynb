```python
# Loading the data and exporting them into a list

from math import *
import matplotlib.pyplot as plt
%matplotlib inline
import statistics
from scipy.signal import savgol_filter
import numpy as np
```

```python
def find_loss(mot,path):
    loss_data=0
    loss_str=0
    loss=[]
    loss_avg=[]    
    liste = []
    
    fh = open(path)
    read_lines = fh.readlines()

    for line in read_lines:
        liste.append(line)
    fh.close()
    
    if mot=='train':
        mot1= 'Epoch: ['
    if mot=='test':
        mot1= 'Test: ['
    idx=0
    for i in range(0,len(liste),1):    
        if mot1 in liste[i]:
            idx+=1
            loss_data=liste[i].split("\t")[-1]
            loss_str=loss_data.split()[-2]

            if float(loss_str)<1000:
                loss.append(float(loss_str))
            #if float(loss_str[1:-2])<1000:
                #loss.append(float(loss_str[1:-2]))
    #print(idx)
    return loss


#LOSS_AVG0.append(sqrt(abs(float(num_list[i][-13:-2]))))
```
